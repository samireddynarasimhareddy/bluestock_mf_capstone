```python
nav_history = pd.read_csv(
    "../data/processed/02_nav_history.csv"
)

nav_history.head()
```

```python
nav_history['date'] = pd.to_datetime(
    nav_history['date']
)

nav_history.head()
```

```python
nav_history['nav'].describe()
```

```python
plt.figure(figsize=(8,5))

plt.hist(
    nav_history['nav'],
    bins=30
)

plt.title("NAV Distribution")
plt.xlabel("NAV")
plt.ylabel("Frequency")

plt.show()
```

```python
print("Start Date:", nav_history['date'].min())
print("End Date:", nav_history['date'].max())
```

```python
monthly_nav = (
    nav_history
    .groupby(
        nav_history['date'].dt.to_period('M')
    )['nav']
    .mean()
)

monthly_nav.head()
```

```python
plt.figure(figsize=(12,5))

monthly_nav.plot()

plt.title("Average Monthly NAV Trend")

plt.xlabel("Month")
plt.ylabel("Average NAV")

plt.show()
```

```python
nav_history.sort_values(
    'nav',
    ascending=False
).head(10)
```

```python
aum = pd.read_csv(
    "../data/processed/03_aum_by_fund_house.csv"
)

aum.head()
```

```python
aum.shape
```

```python
aum.isnull().sum()
```

```python
aum['aum_crore'].describe()
```

```python
top_aum = (
    aum.groupby('fund_house')
       ['aum_crore']
       .max()
       .sort_values(ascending=False)
)

top_aum.head(10)
```

```python
plt.figure(figsize=(12,5))

top_aum.head(10).plot(kind='bar')

plt.title("Top 10 Fund Houses by AUM")

plt.xlabel("Fund House")
plt.ylabel("AUM (Crore)")

plt.xticks(rotation=45)

plt.show()
```

```python
aum.groupby('fund_house')\
['num_schemes']\
.max()\
.sort_values(ascending=False)\
.head(10)
```

```python
plt.figure(figsize=(12,5))

aum.groupby('fund_house')\
['num_schemes']\
.max()\
.sort_values(ascending=False)\
.head(10)\
.plot(kind='bar')

plt.title("Fund Houses by Number of Schemes")

plt.show()
```

```python
aum['date'] = pd.to_datetime(aum['date'])
```

```python
monthly_aum = (
    aum.groupby('date')
       ['aum_crore']
       .sum()
)

monthly_aum.head()
```

```python
plt.figure(figsize=(12,5))

monthly_aum.plot()

plt.title("Industry AUM Trend")

plt.xlabel("Date")
plt.ylabel("Total AUM (Crore)")

plt.show()
```

```python
category = pd.read_csv(
    "../data/processed/05_category_inflows.csv"
)

category.head()
```

```python
category.groupby('category')['net_inflow_crore'].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(10,5))

category.groupby('category')['net_inflow_crore']\
        .sum()\
        .sort_values(ascending=False)\
        .head(10)\
        .plot(kind='bar')

plt.title('Top 10 Categories by Net Inflow')
plt.ylabel('Net Inflow (Crore)')
plt.show()
```

```python
### Insights

- Flexi Cap category attracts significant inflows.
- Mid Cap and Small Cap funds show strong investor interest.
- Large Cap funds remain a stable investment category.
- Category inflows help identify market trends and investor preferences.
```
