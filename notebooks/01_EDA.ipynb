```python
fund_master['launch_date'] = pd.to_datetime(
    fund_master['launch_date']
)

fund_master['launch_year'] = fund_master['launch_date'].dt.year

fund_master['launch_year'].value_counts().sort_index()
```

```python
plt.figure(figsize=(12,5))

fund_master['launch_year'].value_counts()\
.sort_index()\
.plot()

plt.title("Schemes Launch Trend")

plt.show()
```
